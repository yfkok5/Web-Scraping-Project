```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url_2 = "https://www.myer.com.au/c/home/all-appliances?pageNumber=2"
```

```python
# Use the below headers when requesting the url may help you to avoid being blocked from the site
HEADERS = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}

#result = requests.get(url, hea)

#doc = BeautifulSoup(result.text,"html.parser")
```

```python
# Get the HTML content get_url()
def get_url(url):
    try:
        response = requests.get(url,headers = HEADERS).text
        soup = BeautifulSoup(response, 'html.parser')
        return soup
    except Exception as err:
        print('ERROR BY REQUEST:', err)
        return None
```

```python
get_url(url_2).prettify()[0:1000]
```

```python
soup = get_url(url_2)
```

```python
# find all the products on the page: depending on when you run it may return different number, you can cross-ref on the website
products = soup.find_all('li', {'data-automation':'product-grid-item'})
print(len(products))
```

```python
# you pick up the first product, look for the span element and the specific product price currently on display
# i used .text to extract the current price

products[0].find("span" , {"data-automation": "product-price-was"}).text
```

```python
#Creating empty list to be appended on
products_DF = []
price_was = []
price_now = []

for each_item in range(0,len(products)):

    #Extract product_list
    product_item = products[each_item].find("span" , {"data-automation": "product-name"}).text
    products_DF.append(product_item)

    #
    try:
        product_price_was = products[each_item].find("span" , {"data-automation": "product-price-was"}).text
        price_was.append(product_price_was)
    except:
        product_price_was = 'Empty'
        price_was.append('Empty')
    
    try:
        product_price_now = products[each_item].find("span" , {"data-automation": "product-price-now"}).text
        price_now.append(product_price_now)
    except:
        product_price_now = 'Empty'
        price_now.append('Empty')

d = {'Products': products_DF, 'Price Was': price_was, 'Price Now': price_now}

myer_product_df = pd.DataFrame(data=d)
myer_product_df
```

```python
#export to csv
from pathlib import Path  
filepath = Path('..\Web-Scraping-Project\myershoplist.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
myer_product_df.to_csv(filepath)
```
